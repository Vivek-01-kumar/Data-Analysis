### 5.Find the percentage of the top-rated movies that are in: 1 genre only, 2 genres only and 3 genres only

```python
spaces_in_genre1 = df['genre_1'].isna().sum()
spaces_in_genre2 = df['genre_2'].isna().sum()
spaces_in_genre3 = df['genre_3'].isna().sum() 

print(f'total no. of movies with space in genre 1 only : {spaces_in_genre1}')
print(f'total no. of movies with space in genre 1 only : {spaces_in_genre2}')
print(f'total no. of movies with space in genre 1 only : {spaces_in_genre3}')
print('*************************************************************************************')
# If genre 2 has spaces that means it belongs to 1 genre only 
total_in_genre1_only = spaces_in_genre2
print(f" percentage of movies in 1 genre only is : {total_in_genre1_only}")

# If genre 3 with spaces is 26, hten 100 - 26 = 74 records in total have values in genre 3
total_in_genre3_only = 100 - spaces_in_genre3      
print(f" percentage of movies in 3 genre only is : {total_in_genre3_only}")    

# We get genre 2 only percentage finding the remaning percentage after adding genre 1 only and genre 3 only 
total_in_genre2_only = 100 - (total_in_genre1_only + total_in_genre3_only)     
print(f" percentage of movies in 2 genre only is : {total_in_genre2_only}") 
```

### 6.Convert the budget and gross values from ‚Äúdollars‚Äù to ‚Äúdollars in millions‚Äù for all top-rated movies. Round the converted values down to 3 decimal places.

```python
def converter(budget):
    
    return str(round(budget/1000000,3))+'million'

df['budget'] = df.budget.apply(converter)
df['Gross'] = df.Gross.apply(converter) 
df
```

### 7.List all details for the top 10 movies with the highest profit, sorted from highest to lowest.

```python
import pandas as pd 
df = pd.read_csv(r'D:\python assignment\Movie Assignment Data.csv')
# for profit we want to subtract gross from budget 

df['profit'] = df['Gross'] - df['budget']

# we want top ten movies, so we have to sort values in descending order i.e higest to lowest

df.sort_values(by='profit', ascending=False).Title.value_counts().head(10)
```

### 8.Generate a list of all the actors, in alphabetical order by the first name, that have starred in a top-rated movie. If an actor has starred in multiple movies, their name should appear only once on the list.

```python
actor_name1 = df['actor_1_name']
actor_name2 = df['actor_2_name']
actor_name3 = df['actor_3_name']

# merge all the columns of the lead actor,second lead actor and third lead actor in the movie

all_actors_names = pd.concat([actor_name1,actor_name2,actor_name3])

# sort all the actors in alphabetical order, remove duplicates and display their names 

print(f'list of actors that have starred in top-rated movies (in alphabetical order) are:\n{all_actors_names.drop_duplicates().sort_values()}')
```

### 9.The movie studio wants to determine who it should approach to act in its next movie production. Find the top 3 actors who appeared in the most top-rated movies.

```python
actor_1 = df['actor_1_name']
actor_2 = df['actor_2_name']
actor_3 = df['actor_3_name']

# merge all the columns of the lead actor,second lead actor and third lead actor in the movie

movies_actor = pd.concat([actor_1, actor_2, actor_3])

# Top three actors who appeared in the most top-rated movies.

movies_actor.value_counts().head(3)
```

### 10.Create a data visualization that shows each country and the number of top-rated movies produced in it. Find the country that produced the most top-rated movies.

```python
import matplotlib.pyplot as plt
movie_data_visual = df['Country'].value_counts()
movie_data_visual.to_frame()
movie_data_visual.plot(kind='bar' ,y='country', figsize=(20,12))
plt.xlabel('countries')
plt.ylabel('no. of movies')
plt.show()
print(f'country that has produced the most top rated movies is {df["Country"].value_counts().head(1)}')
```
